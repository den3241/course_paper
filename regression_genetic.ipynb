```python
from typing import Tuple, List
import numpy as np
import matplotlib.pyplot as plt
import random

class GeneticRegression():
    def __init__(self,
                population_size:int = 100,
                mutation_rate:int = 0.1,
                mutation_strength:float = 0.5,
                max_generations:int = 100,
                k_range:Tuple[float, float] = (-10, 10),
                b_range:Tuple[float, float] = (-10, 10)):
        self.population_size = population_size
        self.mutation_rate = mutation_rate
        self.mutation_strength = mutation_strength
        self.max_generations = max_generations
        self.k_range = k_range
        self.b_range = b_range
        self.best_k = None
        self.best_b = None


    def _initialize_population(self) -> List[Tuple[float, float]]:
        population = []
        for _ in range(self.population_size):
            k = random.uniform(*self.k_range)
            b = random.uniform(*self.b_range)
            population.append((k, b))
        return population
    
    def _mse(self, y_true:np.ndarray, y_pred:np.ndarray) -> float:
        return np.mean((y_true - y_pred)**2)
    
    def _fitness(self, individ:Tuple[float, float], X:np.ndarray, y:np.ndarray) -> List[Tuple[float, float]]:
        k, b = individ
        y_pred = k*X + b
        return self._mse(y, y_pred)
    
    def _crossover(self, parent1: Tuple[float, float], parent2:Tuple[float, float]) -> Tuple[float, float]:
        alpha = random.random()
        k1, b1 = parent1
        k2, b2 = parent2
        k_child = alpha*k1 + (1-alpha)*k2
        b_child = alpha*b1 + (1-alpha)*b2
        return (k_child, b_child)
    
    def _mutate(self, individ: Tuple[float, float]) -> Tuple[float, float]:
        k, b = individ
        if random.random() < self.mutation_rate:
            k += random.uniform(-self.mutation_strength, self.mutation_strength)
            k = np.clip(k, *self.k_range)
        if random.random() < self.mutation_rate:
            b += random.uniform(-self.mutation_strength, self.mutation_strength)
            b = np.clip(k, *self.b_range)
        return (k, b)

    def predict(self, X):
        if self.best_k == None or self.best_b == None:
            raise ValueError("The model is not trained. Call fit() first")
        return self.best_k * X + self.best_b
    
    def _selection(self, population: List[Tuple[float, float]]) -> List[Tuple[float, float]]:
        selected = dict()
    
    def fit():
        pass
            
# def _selection(self, population: List[Tuple[float, float]],
#                fitnesses: List[float]) -> List[Tuple[float, float]]:
#         selected = []
        
#         # Элитизм: сохраняем лучших особей
#         elite_indices = np.argsort(fitnesses)[:self.elitism_count]
#         for idx in elite_indices:
#             selected.append(population[idx])
        
#         # Турнирная селекция для остальных
#         tournament_size = max(2, self.population_size // 10)
        
#         while len(selected) < self.population_size:
#             # Выбираем случайных особей для турнира
#             tournament_indices = random.sample(range(len(population)), tournament_size)
#             tournament_fitness = [fitnesses[i] for i in tournament_indices]
            
#             # Выбираем лучшую (с наименьшей MSE)
#             winner_idx = tournament_indices[np.argmin(tournament_fitness)]
#             selected.append(population[winner_idx])
        
#         return selected

# def fit(self, X: np.ndarray, y: np.ndarray, 
#             verbose: bool = True, 
#             early_stopping_patience: int = 10) -> 'GeneticRegression':
#         """
#         Обучение модели с помощью генетического алгоритма
        
#         Parameters:
#         -----------
#         X : np.ndarray
#             Входные данные
#         y : np.ndarray
#             Целевые значения
#         verbose : bool
#             Выводить ли информацию о процессе обучения
#         early_stopping_patience : int
#             Количество поколений без улучшений для ранней остановки
            
#         Returns:
#         --------
#         self
#             Обученная модель
#         """
#         # Инициализация популяции
#         population = self._initialize_population()
        
#         best_fitness_generation = float('inf')
#         patience_counter = 0
        
#         for generation in range(self.max_generations):
#             # Вычисление приспособленности для каждой особи
#             fitnesses = [self._fitness(ind, X, y) for ind in population]
            
#             # Обновление лучшей особи
#             min_fitness_idx = np.argmin(fitnesses)
#             current_best_fitness = fitnesses[min_fitness_idx]
#             current_best_individual = population[min_fitness_idx]
            
#             if current_best_fitness < self.best_fitness:
#                 self.best_fitness = current_best_fitness
#                 self.best_k, self.best_b = current_best_individual
#                 patience_counter = 0
#             else:
#                 patience_counter += 1
            
#             self.fitness_history.append(self.best_fitness)
            
#             # Проверка сходимости
#             if generation > 0:
#                 fitness_improvement = abs(self.fitness_history[-2] - self.fitness_history[-1])
#                 if fitness_improvement < self.convergence_threshold:
#                     if verbose:
#                         print(f"Сходимость достигнута на поколении {generation}")
#                     break
            
#             # Ранняя остановка
#             if patience_counter >= early_stopping_patience:
#                 if verbose:
#                     print(f"Ранняя остановка на поколении {generation}")
#                 break
            
#             # Вывод информации
#             if verbose and generation % 10 == 0:
#                 print(f"Поколение {generation}: Лучшая MSE = {self.best_fitness:.6f}")
            
#             # Селекция
#             selected = self._selection(population, fitnesses)
            
#             # Скрещивание и мутация
#             new_population = []
            
#             # Сохраняем элитных особей
#             for i in range(self.elitism_count):
#                 new_population.append(selected[i])
            
#             # Генерация нового поколения
#             while len(new_population) < self.population_size:
#                 # Выбираем двух родителей
#                 parent1, parent2 = random.sample(selected, 2)
                
#                 # Скрещивание
#                 child = self._crossover(parent1, parent2)
                
#                 # Мутация
#                 child = self._mutate(child)
                
#                 new_population.append(child)
            
#             population = new_population
        
#         if verbose:
#             print(f"\nОбучение завершено.")
#             print(f"Лучшие параметры: k = {self.best_k:.6f}, b = {self.best_b:.6f}")
#             print(f"Лучшая MSE: {self.best_fitness:.6f}")
        
#         return self
```
